```python
import pandas as pd
```

```python
from sklearn import preprocessing
```

```python
import numpy as np
```

```python
import tensorflow as tf
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/diabetes.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
lab=preprocessing.LabelEncoder()
```

```python
df['Result']=lab.fit_transform(df["Result"])
```

```python
df["Result"].unique()
```

```python
x=df[["Pregnancies","Glucose","BloodPressure","SkinThickness","Insulin","BMI","DiabetesPedigreeFunction","Age"]]
```

```python
x
```

```python
y=df["Result"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.2,random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
ann=tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=12,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=12,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss="binary_crossentropy",metrics=["accuracy"])
```

```python
ann.fit(X_train,Y_train,batch_size=64,epochs=500)
```

```python
ann.fit(X_test,Y_test,batch_size=64,epochs=500)
```

```python
print(ann.predict(sc.transform([[1,93,70,31,0,30.4,0.315,23]]))>0.5)
```

```python
print(ann.predict(sc.transform([[0,137,40,35,168,43.1,2.288,33]]))>0.5)
```
